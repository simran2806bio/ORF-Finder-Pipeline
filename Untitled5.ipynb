```python
# Full ORF pipeline for multi_gene_sample.fasta
!pip install biopython --quiet
from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
import re
import csv
import os
import pandas as pd
import matplotlib.pyplot as plt

# Parameters
FASTA_IN = "multi_gene_sample.fasta"
OUT_FASTA = "multi_gene_proteins.fasta"
OUT_CSV = "multi_gene_orfs.csv"
MIN_AA = 20   # change this to 30/20 if you want laxer results

# Utility functions
def six_frame_translations_trim(dna_seq):
    seq = Seq(str(dna_seq))
    trimmed_len = (len(seq)//3) * 3
    trimmed = seq[:trimmed_len]
    frames = []
    # forward
    for offset in range(3):
        frames.append((offset, str(trimmed[offset:].translate()), offset, '+', trimmed))
    # reverse complement
    rc = trimmed.reverse_complement()
    for offset in range(3):
        frames.append((3+offset, str(rc[offset:].translate()), offset, '-', trimmed))
    return frames, trimmed_len

def find_orfs_from_protein_with_index(protein_seq, min_aa=1):
    orfs = []
    segments = protein_seq.split('*')
    aa_index = 0
    for seg in segments:
        for m in re.finditer('M', seg):
            start = aa_index + m.start()
            orf = seg[m.start():]
            if len(orf) >= min_aa:
                orfs.append((start, orf))
        aa_index += len(seg) + 1
    return orfs

# Process all records
records = list(SeqIO.parse(FASTA_IN, "fasta"))
orf_records = []
csv_rows = []

for rec in records:
    frames_info, trimmed_len = six_frame_translations_trim(rec.seq)
    # for each frame, find ORFs >= MIN_AA
    for frame_idx, prot, frame_offset, strand, trimmed in frames_info:
        orfs = find_orfs_from_protein_with_index(prot, min_aa=MIN_AA)
        for idx, (aa_start_idx, aa_seq) in enumerate(orfs):
            aa_len = len(aa_seq)
            # compute nucleotide coordinates (1-based) depending on strand
            if strand == '+':
                nt_start_0 = frame_offset + aa_start_idx*3
                nt_end_0 = frame_offset + (aa_start_idx + aa_len)*3 - 1
                nt_start = nt_start_0 + 1
                nt_end = nt_end_0 + 1
            else: # strand == '-'
                nt_start_0_rev = frame_offset + aa_start_idx*3
                nt_end_0_rev = frame_offset + (aa_start_idx + aa_len)*3 - 1
                nt_start = trimmed_len - nt_end_0_rev
                nt_end = trimmed_len - nt_start_0_rev

            orf_id = f"{rec.id}_frame{frame_idx}_orf{idx}_len{aa_len}"
            orf_record = SeqRecord(Seq(aa_seq), id=orf_id, description=f"nucleotide_range=[{nt_start}-{nt_end}]_strand={strand}_frame_offset={frame_offset}")
            orf_records.append(orf_record)
            csv_rows.append([rec.id, orf_id, aa_seq, frame_idx, frame_offset, strand, nt_start, nt_end, aa_len])

# Write output FASTA and CSV
SeqIO.write(orf_records, OUT_FASTA, "fasta")

csv_header = ["record_id", "orf_id", "protein_seq", "frame_index", "frame_offset", "strand", "nuc_start", "nuc_end", "aa_length"]
with open(OUT_CSV, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(csv_header)
    writer.writerows(csv_rows)

print(f"ORFs written to {OUT_FASTA} and {OUT_CSV}")

# Display CSV as a DataFrame
df_orfs = pd.read_csv(OUT_CSV)
display(df_orfs)

# Visualize protein length distribution
plt.figure(figsize=(10, 6))
plt.hist(df_orfs['aa_length'], bins=20, edgecolor='black')
plt.title('Distribution of ORF Protein Lengths')
plt.xlabel('Protein Length (amino acids)')
plt.ylabel('Number of ORFs')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
